```python
import pandas as pd
import numpy as np
import pandas.io.json as js
import re as re
```

```python
raw_data = pd.read_json('json_data/react.json')
react_vacancies = js.json_normalize(raw_data['items'])

pdreact = react_vacancies[[
        'name',
        'salary.currency',
        'salary.from',
        'salary.to',
        'snippet.requirement',
        'snippet.responsibility'
    ]]

KEY_1 = 'name'
KEY_2 = 'snippet.responsibility'
KEY_3 = 'snippet.requirement'
```

```python
# -- 1. by str.contains method ---

# nameWords = ['Front']    
# pattern = '|'.join(map(re.escape, nameWords))

# nameCondition = pdreact['name'].str.contains(pattern, case=False)
# reqCondition = pdreact['snippet.requirement'].str.contains(pattern, case=False)
# resCondition = pdreact['snippet.responsibility'].str.contains('front', case=False)

# filteredResult = pdreact[nameCondition]
```

```python
# -- 2. by filter method ---

pattern = re.compile('фронт|front', re.IGNORECASE)

nameCondition = pdreact[KEY_1].str.contains(pattern)
resCondition = pdreact[KEY_2].str.contains(pattern).fillna(False)
reqCondition = pdreact[KEY_3].str.contains(pattern).fillna(False)

conditions = nameCondition | resCondition | reqCondition

filteredResult = pdreact[resCondition]
```
